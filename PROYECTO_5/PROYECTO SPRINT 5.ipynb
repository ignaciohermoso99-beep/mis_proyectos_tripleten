## Agregar datos por usuario


```python
# Calcula el número de llamadas hechas por cada usuario al mes. Guarda el resultado.

monthly_calls_per_user = calls.groupby(['user_id','month']).agg(
    calls_count=('duration','count')
).reset_index()

monthly_calls_per_user.head(5)
```

```python
# Calcula la cantidad de minutos usados por cada usuario al mes. Guarda el resultado.
monthly_minutes_per_user = calls.groupby(["user_id","month"]).agg(
    minutes_used=("duration","sum")
).reset_index()

monthly_minutes_per_user.head()
```

<div class="alert alert-block alert-success">
<b>Comentario de Revisor            </b> <a class="tocSkip"></a>

Muy bien, correcto el cálculo del consumo mensual de llamadas para cada usuario.


</div>

```python
# Calcula el número de mensajes enviados por cada usuario al mes. Guarda el resultado.
monthly_messages_per_user = messages.groupby(["user_id","month"]).agg(
    sent_messages=("id","count")
).reset_index()
monthly_messages_per_user.head()

```

```python
# Calcula el volumen del tráfico de Internet usado por cada usuario al mes. Guarda el resultado.
monthly_mb_used_per_user = internet.groupby(["user_id","month"]).agg(
    mb_used_total = ("mb_used","sum")
).reset_index()
monthly_mb_used_per_user.head(5)
```

<div class="alert alert-block alert-success">
<b>Comentario de Revisor            </b> <a class="tocSkip"></a>

Bien, correcto el cálculo de consumo de mensajes e internet

</div>

```python
# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month
monthly_usage = monthly_calls_per_user.merge(
    monthly_minutes_per_user,
    on = ["user_id","month"],
    how = "outer"
)
monthly_usage = monthly_usage.merge(
    monthly_messages_per_user,
    on = ["user_id","month"],
    how = "outer"    
)
monthly_usage = monthly_usage.merge(
    monthly_mb_used_per_user,
    on = ["user_id","month"],
    how = "outer"    
)
monthly_usage.head()
```

<div class="alert alert-block alert-success">
<b>Comentario de Revisor     </b> <a class="tocSkip"></a>

Correcto el uso del merge tipo outer, esto evita que se pierdan datos de usuarios.

</div>

```python
# Añade la información de la tarifa
# Unimos monthly_usage con users para traer la info del plan de cada user
monthly_usage = monthly_usage.merge(
    users[['user_id', 'plan']],
    on='user_id',
    how='left'
)

# 2. Ahora unimos con plans para traer la tarifa mensual y demás info del plan y tener un df con info útil y sin columnas duplicadas
monthly_usage = monthly_usage.merge(
    plans[['plan_name', 'usd_monthly_fee', 'minutes_included', 
           'messages_included', 'mb_per_month_included',
           'usd_per_minute', 'usd_per_message', 'usd_per_gb']],
    left_on='plan',
    right_on='plan_name',

    how='left'

)

monthly_usage = monthly_usage.drop(columns=['plan_name'])

monthly_usage.head(30)
```

<div class="alert alert-block alert-success">
<b>Comentario de Revisor            </b> <a class="tocSkip"></a>

Bien, en esta parte basta con el uso de left ya que la data de consumo de usuarios ya está completa.

</div>

```python
# Calcula el ingreso mensual para cada usuario

monthly_usage = monthly_usage.fillna(0)    


monthly_usage["extra_minutes"] = (monthly_usage["minutes_used"] - monthly_usage["minutes_included"]).clip(lower = 0)
monthly_usage["extra_messages"] = (monthly_usage["sent_messages"] - monthly_usage["messages_included"]).clip(lower = 0)
monthly_usage["extra_mb"] = (monthly_usage["mb_used_total"] - monthly_usage["mb_per_month_included"]).clip(lower = 0)

monthly_usage["extra_gb"] = np.ceil(monthly_usage["extra_mb"] / 1024)

monthly_usage["cost_minutes"] = monthly_usage["extra_minutes"] * monthly_usage["usd_per_minute"] 
monthly_usage["cost_messages"] = monthly_usage["extra_messages"] * monthly_usage["usd_per_message"] 
monthly_usage["cost_extra_gb"] = monthly_usage["extra_gb"] * monthly_usage["usd_per_gb"]

monthly_usage["total_revenue"] = (
    monthly_usage["usd_monthly_fee"] +
    monthly_usage["cost_minutes"] + 
    monthly_usage["cost_messages"] + 
    monthly_usage["cost_extra_gb"] 
)

monthly_usage.head(30)
```

# Prueba las hipótesis
(H₀): Los ingresos promedio de los usuarios del plan Ultimate son diferentes a los del plan Surf.
(H₁): Los ingresos promedio de los usuarios de ambos planes son iguales.

[Prueba la hipótesis de que el ingreso promedio de los usuarios del área NY-NJ es diferente al de los usuarios de otras regiones.]

```python

from scipy.stats import ttest_ind

ultimate_revenue = monthly_usage.loc[monthly_usage['plan']=='ultimate', 'total_revenue']
surf_revenue = monthly_usage.loc[monthly_usage['plan']=='surf', 'total_revenue']


t_stat, p_value = ttest_ind(ultimate_revenue, surf_revenue, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Rechazamos H₀: los ingresos promedio son diferentes")
else:
    print("No rechazamos H₀: no hay evidencia suficiente para decir que los ingresos promedio difieren")
```
